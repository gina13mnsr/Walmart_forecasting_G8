**EDA**

*1 ) Which store has maximum sales ?*

```python
# Agrupar por tienda y sumar las ventas
sales_by_store = df.groupby("Store")[["Weekly_Sales"]].sum()

# Ordenar las ventas de forma descendente
sales_by_store = sales_by_store.sort_values(by="Weekly_Sales", ascending=False)

sales_by_store
```

```python
# Mostrar la tienda con más ventas
print("La tienda con más ventas es la número:", sales_by_store.index[0])
```

```python
# visualizar de manera grafica el resultado
sales_by_store.plot(kind="bar", figsize=(12,6), legend=None)
plt.title("Ventas por tienda")
plt.xlabel("Tienda")
plt.ylabel("Ventas")

# Mostrar la tienda con más ventas
print("La tienda con más ventas es la número:", sales_by_store.index[0])

# Mostrar el gráfico
plt.show()
```

**2) Qué tienda tiene la desviación estándar máxima, es decir, las ventas varían mucho. Además, averigüe el coeficiente de desviación medio**

```python
# Agrupar por tienda y calcular la desviación estándar y la media
sales_std_mean = df.groupby("Store")[["Weekly_Sales"]].agg(['std', 'mean'])

# Calcular el coeficiente de desviación para cada tienda
sales_std_mean["coef_of_var"] = sales_std_mean[("Weekly_Sales", "std")] / sales_std_mean[("Weekly_Sales", "mean")]

# Ordenar las tiendas por coeficiente de desviación de forma descendente
sales_std_mean = sales_std_mean.sort_values(by="coef_of_var", ascending=False)

# Mostrar la tienda con mayor coeficiente de desviación
print("La tienda con mayor coeficiente de desviación es la número:", sales_std_mean.index[0])

# Calcular el coeficiente de desviación medio para todas las tiendas
coef_of_var_mean = sales_std_mean["coef_of_var"].mean()

# Mostrar el coeficiente de desviación medio
print("El coeficiente de desviación medio es:", coef_of_var_mean)
```

```python
# Crear el gráfico de puntos
plt.scatter(sales_std_mean.index, sales_std_mean["coef_of_var"])
plt.title("Coeficiente de desviación por tienda")
plt.xlabel("Tienda")
plt.ylabel("Coeficiente de desviación")

# Mostrar el gráfico
plt.show()
```

En el grafico anterior observamos que la tienda 35 destaca notablemente por poseer un coeficiente de desviacion significatibvamente mayor de los demas, esta tienda 35 tiene patrones de venta impredecibles (volatiles) al igual de otras tiendas que destacan en el grafico como puntos mas altos, de lo que se concluye que estas tiendas como la 35 requieren atencion especial en terminos de planificacion y estrategia de ventas.

**3) ¿Qué tienda/s tiene una buena tasa de crecimiento trimestral en el tercer trimestre de 2012?**

```python
# Splitting la columna dato en columna 'day', 'month', 'year'
df["Day"]= pd.DatetimeIndex(df['Date']).day
df['Month'] = pd.DatetimeIndex(df['Date']).month
df['Year'] = pd.DatetimeIndex(df['Date']).year
df
```

```python
# Filtrar los datos para incluir solo el tercer trimestre de 2012
q3_2012 = df[(df["Date"].dt.year == 2012) & (df["Date"].dt.month.isin([7, 8, 9]))]

# Agrupar los datos por tienda y calcular la tasa de crecimiento trimestral
sales_growth = q3_2012.groupby("Store")[["Weekly_Sales"]].agg(lambda x: (x.iloc[-1] / x.iloc[0]) - 1)

# Ordenar los datos por la tasa de crecimiento trimestral de forma descendente
sales_growth = sales_growth.sort_values(by="Weekly_Sales", ascending=False)

# Seleccionar las tiendas con la tasa de crecimiento trimestral más alta
best_stores = sales_growth[sales_growth["Weekly_Sales"] > 0]

# Mostrar las tiendas con la tasa de crecimiento trimestral más alta
print("Las tiendas con la tasa de crecimiento trimestral más alta son:")
print(best_stores)
```

EL crecimiento trimestral mas alto en ventas semanales del tercer trimestre del 2012 lo tuvo la tienda 44 con un crecimiento del 17.09% en comparacion con las ventas  del primer trimestre del 2012.
